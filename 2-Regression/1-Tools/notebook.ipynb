```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection

linnerud = datasets.load_linnerud()
X = linnerud.data
y = linnerud.target
w_line= X[:,1]
print(w_line)
s_line= y[:,1]
print(s_line)

w_line_train, w_line_test, s_line_train, s_line_test = model_selection.train_test_split(w_line, s_line, test_size=0.33)

model = linear_model.LinearRegression()
model.fit(w_line_train.reshape(-1, 1), s_line_train)

s_line_pred = model.predict(w_line_test.reshape(-1, 1))
plt.scatter(w_line_test, s_line_test,  color='black')
plt.plot(w_line_test, s_line_pred, color='blue', linewidth=3)
```
